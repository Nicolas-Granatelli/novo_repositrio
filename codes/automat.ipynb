```python
import pandas as pd
import numpy as np
import os
```

```python
base = pd.read_excel("./dados.xlsx")
base
```

```python
dados_modificado = base.copy()
dados_modificado['Genero'] = np.where(dados_modificado['Genero'] == 'F', "Feminino", "Masculino")
dados_modificado


```

```python
from datetime import datetime

dados_modificado['Idade'] = pd.Series()
hoje = datetime.now()
for i in range(len(dados_modificado)):
    dados_modificado['Idade'][i] = (hoje - dados_modificado['Nascimento'][i]).days // 365
dados_modificado['Idade'] = dados_modificado['Idade'].astype(int)
dados_modificado

def gerar_email(nome):
    texto = nome.split()
    primeiro = texto[0]
    ultimo = texto[-1]
    return (primeiro + "." + ultimo + "@germinare.org.br").lower()


dados_modificado['Email'] = pd.Series()
for i in range(len(dados_modificado)):
    dados_modificado['Email'][i] = gerar_email(dados_modificado['Nome'][i])
dados_modificado
```

```python
dados_modificado = dados_modificado.drop('N', axis=1)
dados_modificado
```

```python
dados_modificado.to_excel("Finalizado.xlsx", index=False)
```

```python
base_pdf = pd.read_excel("./Finalizado.xlsx", usecols= ['Nome', 'Idade', 'Jantar', 'Turma', 'Sala', 'Dia', 'Email'])
base_pdf
```

```python
from fpdf import FPDF


for i in range(len(base_pdf)):
    X = (f"Nome: {base_pdf['Nome'][i]}", f"Idade: {base_pdf['Idade'][i]}", f"Vai Jantar?: {base_pdf['Jantar'][i]}", f"Turma: {base_pdf['Turma'][i]}" , f"Sala: {base_pdf['Sala'][i]}" , f"Dia: {base_pdf['Dia'][i]}" , f"Email: {base_pdf['Email'][i]}" )

    name = base_pdf['Nome'][i]
    name_format = name.replace(" ", "_")
    folder = os.path.join("C:\\Users\\Aluno Temp\\OneDrive - Instituto Germinare\\Área de Trabalho\\2 Tech\\Análise de Dados\\Pandas\\Aula 8\\pasta1", name_format)

    os.mkdir(folder)

    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=20)
    for j  in range(len(X)):
        pdf.cell(300,10,txt=str(X[j]), ln=3, align='L')
    pdf.output(f"{folder}\{name_format}.pdf")
```
